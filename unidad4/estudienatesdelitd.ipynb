```python
import scipy.stats as stats

# Datos del problema
media = 70
desviacion = 4
n_estudiantes = 500

# Intervalo de interés
limite_inferior = 68
limite_superior = 72

# Calcular valores Z
z_inferior = (limite_inferior - media) / desviacion
z_superior = (limite_superior - media) / desviacion

# Calcular probabilidades usando la función de distribución acumulada (CDF)
probabilidad = stats.norm.cdf(z_superior) - stats.norm.cdf(z_inferior)

# Calcular número de estudiantes
cantidad_estudiantes = probabilidad * n_estudiantes

# Mostrar resultados
print(f"Probabilidad de pesar entre 68 y 72 kg: {probabilidad:.4f}")
print(f"Cantidad estimada de estudiantes: {cantidad_estudiantes:.0f}")
```
