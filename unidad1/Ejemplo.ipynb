```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
covid = pd.read_csv("20-06-30-COVID19MEXICO - codificado.csv")
```

```python
covid.head()
```

```python
covid.info()
#mostrar la informacion del archivo
```

```python
# mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
# mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
#mostrar los estadisticos descriptivos
covid.describe()
```

```python
media=covid.EDAD.mean()

plt.pyplot.hist(covid.EDAD, bins=11)
plt.pyplot.title('Histograma de Edad')
plt.pyplot.axvline(media, color='y' , linestyle='dashed' , linewidth=2, label = f'Media={media:.2f}')
plt.pyplot.xlabel("Edad")
plt.pyplot.ylabel("Frecuencia")
pyplot.show()
```

```python
plt.pyplot.hist(covid.SEXO, bins=3)
plt.pyplot.title('Histograma de Sexo')
plt.pyplot.xlabel("Sexo")
plt.pyplot.ylabel("Frecuencia")
pyplot.show()
```

```python
plt.pyplot.hist(covid.TIPO_PACIENTE, bins=2)
plt.pyplot.title('Histograma de TIPOS DE PASIENTES')
plt.pyplot.xlabel("TIPOS DE PACIENTES")
plt.pyplot.ylabel("Frecuencia")
pyplot.show()
```

```python
sns.boxplot(x = "SEXO" , y ="EDAD", data=covid)
```
